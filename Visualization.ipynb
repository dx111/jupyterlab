```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

### SDD_TYPE1可视化展示

```python
sdd_type1=pd.read_csv("sdd_type1_result.csv")
sdd_type1.head()
```

```python
sdd_type1=sdd_type1[sdd_type1["lambda"]<1]
sdd_type1["sp_sum"]=(sdd_type1["sp1"]+sdd_type1["sp2"]+sdd_type1["sp3"]+sdd_type1["sp4"])/4.0
fig1=plt.figure()
ax1=Axes3D(fig1)
ax1.scatter(sdd_type1["lambda"],sdd_type1["sigma"],sdd_type1["val_acc"])
ax1.set_xlabel("lambda")
ax1.set_ylabel("sigma")
ax1.set_zlabel("acc")
```
