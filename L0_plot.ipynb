```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
plt.style.use('ggplot')
```

# display regularization function

### define 

```python
def type1(x,sigma=0.05):
    numerator=x**2
    denominator=x**2+sigma**2
    regularization=np.divide(numerator,denominator)
    return regularization
def type2(x,sigma=0.05):
    numerator=(x**2)**2
    denominator=sigma**2
    regularization=1-np.exp(-np.divide(numerator,denominator))
    return regularization
def type3(x,sigma=0.05):
    numerator=1-np.exp(-2*pow(x**2,sigma))
    denominator=1+np.exp(-2*pow(x**2,sigma))
    regularization=np.abs(np.divide(numerator,denominator))
    return regularization
def type4(x,sigma=0.05):
    regularization=pow(x**2,sigma)
    return regularization
```

```python
x=np.linspace(-1,1,100000)
```

## L0 type1
### $\mathrm{L}_{2}=\sum_{j=1}^{N} w_{i j}^{2}$
### $f_{\sigma}(W)=\lambda * \sum_{i=1}^{M} \frac{\mathrm{L}_{2}^{2}}{\mathrm{L}_{2}^{2}+\sigma^{2}}$

```python
y1=type1(x)
plt.plot(x,y1)
```

## L0 type2
### $\mathrm{L}_{2}=\sum_{j=1}^{N} w_{i j}^{2}$
### $f_{\sigma}(W)=\lambda\sum_{i=1}^{M}\left[1-\exp \left(-\frac{\mathrm{L}_{2}^{2}}{2 \sigma^{2}}\right)\right]$

```python
y2=type2(x)
plt.plot(x,y2)
```

## L0 type3
### $\mathrm{L}_{2}=\sum_{j=1}^{N} w_{i j}^{2}$
### $f_{\sigma}(W)=\lambda \sum_{i=1}^{M}\left|\frac{1-\mathrm{e}^{-2 \mathrm{L}_{2}}}{1+\mathrm{e}^{-2 \mathrm{L}_{2}}}\right|$

```python
y3=type3(x)
plt.plot(x,y3)
```

## L0 type4
### $\mathrm{L}_{2}=\sum_{j=1}^{N} w_{i j}^{2}$
### $f_{\sigma}(W)=\lambda\sum_{i=1}^{M} L_{2}^{\sigma}$

```python
y4=type4(x)
plt.plot(x,y4)
```

```python
# L0,p

x=np.linspace(-1,1,100000)
p_=np.arange(1/8,2,1/8)
fig2=plt.figure()

ax=Axes3D(fig2)
for p in p_:
    z=type4(x,p)
    p_array=np.full_like(x,p)
    ax.plot(x,p_array,z)
    ax.set_xlabel("p")
    ax.set_ylabel("x")
    ax.set_zlabel("z")
```
