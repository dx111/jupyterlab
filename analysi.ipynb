```python
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
data=pd.read_csv("result.csv")
data.head()
```

```python
fig1=plt.figure()
fig2=plt.figure()
ax=Axes3D(fig1)
data["sp_sum"]=(data["sp1"]+data["sp2"]+data["sp3"]+data["sp4"])/4.0
ax2=Axes3D(fig2)
data=data[data["lambda1"]<0.5]
sigma=data["sigma"].drop_duplicates().tolist()

for sig in sigma:
    temp=data[data["sigma"]==sig]
    ax.plot(temp["lambda1"],temp["sigma"],temp["val_acc"])
    ax.set_xlabel("lambda")
    ax.set_ylabel("sigma")
    ax.set_zlabel("acc")
    
    ax2.plot(temp["lambda1"],temp["sigma"],temp["sp_sum"])
    ax2.set_xlabel("lambda")
    ax2.set_ylabel("sigma")
    ax2.set_zlabel("sp")
    
```
